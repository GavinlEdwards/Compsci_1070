```python
def number(a:int): 
    count = 1;
    while (count<=a):
        if(count%3==0 and count%5==0):
            print(count,'fizzbuzz');
        elif(count%3==0):
            print(count,'fizz');
        elif(count%5==0):
            print(count,'buzz');
        else:
            print(count)
        count+=1;
number(78)                    
        
    
```

```python
import pandas as pd 
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
from sklearn.preprocessing import StandardScaler

diabetes_df = pd.read_csv("diabetes.csv")

diabetes_df.head()
```

```python
from sklearn.model_selection import train_test_split
X = diabetes_df.drop('Outcome', axis=1)
y = diabetes_df['Outcome']

#split data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=32,stratify = y)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=32)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
```

```python
# decision tree classifier
# deep and complex trees tend to overfit, so we want to generalize and have
# reasonably simple trees as much as possible
model = tree.DecisionTreeClassifier(criterion='gini',max_depth = 5,min_impurity_decrease=.005, random_state=32)
```

```python
model = model.fit(X_resampled,y_resampled)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
tree.plot_tree(model, feature_names = list(X.columns), filled=True, fontsize=6, rounded = True)
plt.show()
plt.savefig("tree.svg")
```
