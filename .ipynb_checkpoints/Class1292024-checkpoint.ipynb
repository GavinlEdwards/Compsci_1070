```python
#default/basic constructor
class a_sample_class:
    
    def __init__(self):
        self.var = 0

        #an instance method for printing
    def print_vars(self):
        print(self.var)


obj = a_sample_class()
obj.print_vars()
```

```python
class a_sample_class:
    #parameterized constructor
    def __init__(self, one, two):
        self.first = one
        
        #an instance method for printing
    def print_vars(self):
        print(self.first)


obj = a_sample_class(3,2)
obj.print_vars()
```

```python
class a_sample_class:
    class_attr = 5
    #parameterized constructor
    def __init__(self, one, two):
        self.first = one
        self.second = two
        
        #an instance method for printing
    def print_vars(self):
        print(self.first)


obj = a_sample_class(17,42)
print(obj.first)
print(obj.class_attr)
print(obj.second)
#print(a_sample_class.first) Doesn't print as first isn't defined as a class variable.
print(a_sample_class.class_attr)
```

# Instance Attributes and Class Attributes
2 scopes for attributes

instance attribute = variable that belongs to a specific instance or instantiation. Defined inside the contructor (init). Can access from the scope of an object (local)

Class attribute = type of variable that belongs to the class itself. 
Defined outside the constructor. Similar to static attributes in Java (not exactly the same). Both a property of a class AND a property of the object. We can mutate a class attribute to be an instance attribute.

```python
#Instance, Class, and Static Methods
class candy:
    def __init__(self,
                 _brand,
                 _is_caramel=False,
                 _has_nuts=False,
                 _calories=30,
                 _type='chocolate',
                 _size='small'):
        self.type=_type
        self.brand = _brand
        self.is_caramel = _is_caramel
        self.calories = _calories
        self.size = _size
    
    #instance method - can modify all the states both the class and instance state printing acceses the state
    def display(self):
        print(f'My favorite candy is {self.brand} {self.type}')

    # class method - used to modify the state of a class (for all instances)
    @classmethod
    def snickers(cls):
        return cls('Snickers',_is_caramel=True,_has_nuts=True)
    # static method - used for comparison or validation
    # commonly used in data for checking if data meets conditions
    # if you want to check the value without needing an instance of it
    @staticmethod
    def calories(size,n):
        if size == 'Large':
            calories = 500
        elif size == 'regular':
            calories = 350
        elif size == 'small':
            calories = 100
        else:
            calories = 0
        calories = calories*n
        return calories
```

```python
favorite_candy = candy(_brand ="Reeses")
favorite_candy.display()
```

```python
#Creates a snickers bar instance
snickers_bar = candy.snickers()
snickers_bar.display()
candy.calories('small',3)
```
