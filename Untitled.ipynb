```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
```

```python
credit_df = pd.read_csv("Credit_card.csv")
accept = pd.read_csv('Credit_card_label.csv')
credit_df.head()
in_or_not = accept['label'].to_numpy()
credit_df['label'] = pd.Series(in_or_not, index=credit_df.index)
credit_df.head(100)
```

```python
credit_df.head()
credit_df['GENDER'] = credit_df['GENDER'].fillna('M')
credit_df['GENDER'] = np.where(credit_df['GENDER'] == 'M',0,np.where(credit_df['GENDER'] == 'F',1,2))
def Y_or_N(name):
    credit_df[name] = credit_df[name].fillna('N')
    credit_df[name] = np.where(credit_df[name] == 'Y',1,np.where(credit_df[name] == 'N',0,2))

Y_or_N('Car_Owner')
Y_or_N('Propert_Owner')
credit_df['Annual_income'] = credit_df['Annual_income'].fillna(credit_df['Annual_income'].median())

credit_df['Type_Income'] = credit_df['Type_Income'].fillna(0)
credit_df['Type_Income'] = np.where(credit_df['Type_Income'] == 'Pensioner',1,np.where(credit_df['Type_Income'] == 'Commercial associate',2,np.where(credit_df['Type_Income'] == 'State servant',3,np.where(credit_df['Type_Income'] == 'Working',4,5))))
credit_df['EDUCATION'] = np.where(credit_df['EDUCATION'] == 'Higher education',0,np.where(credit_df['EDUCATION'] == 'Secondary / secondary special',2,np.where(credit_df['EDUCATION'] == 'Lower secondary',3,np.where(credit_df['EDUCATION'] == 'Incomplete higher',4,np.where(credit_df['EDUCATION'] == 'Academic degree',1,5)))))
credit_df['Marital_status'] = np.where(credit_df['Marital_status'] == 'Married',0,np.where(credit_df['Marital_status'] == 'Civil marriage',1,np.where(credit_df['Marital_status'] == 'Widow',2,np.where(credit_df['Marital_status'] == 'Separated',3,np.where(credit_df['Marital_status'] == 'Single / not married',4,5)))))
credit_df['Housing_type'] = np.where(credit_df['Housing_type'] == 'House / apartment',0,np.where(credit_df['Housing_type'] == 'Office apartment',1,np.where(credit_df['Housing_type'] == 'Co-op apartment',2,np.where(credit_df['Housing_type'] == 'Municipal apartment',3,np.where(credit_df['Housing_type'] == 'Rented apartment',4,np.where(credit_df['Housing_type'] == 'With parents',5,6))))))
credit_df['Employed_days'] = np.where(credit_df['Employed_days'] < 0,0,1) 
# Cleaned up employed days to be included 
```
