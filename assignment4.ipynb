```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

survey_results_df = pd.read_csv('CSCI_1070_SURVEY_RESULTS.csv')
survey_results_df.columns = survey_results_df.columns.str.replace(' ','')
survey_results_df
```

```python
survey_results_df.columns
survey_results_df['Whatisyourname?'].dtype
```

```python
survey_results_df['Whatisyourname?'] = survey_results_df['Whatisyourname?'].str.replace(' ','_')
survey_results_df['Whatstate/countrywereyoubornin?'] = survey_results_df['Whatstate/countrywereyoubornin?'].str.replace(', ','-').str.replace('/','-')
survey_results_df['Whatstate/countrywereyoubornin?'] = survey_results_df['Whatstate/countrywereyoubornin?'].str.replace(' ','_')
survey_results_df['Whatstate/countrywereyoubornin?'] = survey_results_df['Whatstate/countrywereyoubornin?'].str.replace('U.S','USA').str.replace('USA','United_States')
survey_results_df['Howoldareyou?'] = survey_results_df['Howoldareyou?'].str.replace(' ','').str.replace('18yearsold','18').str.replace('19turning20onFebruary28','19')
survey_results_df['Howoldareyou?'] = survey_results_df['Howoldareyou?'].fillna('18')
survey_results_df['Howoldareyou?'] = pd.to_numeric(survey_results_df['Howoldareyou?'], errors ='coerce', downcast='integer')
survey_results_df['Howmanypetsandofwhichkinddoyouhave?'] = survey_results_df['Howmanypetsandofwhichkinddoyouhave?'].str.replace('none','0')
survey_results_df['Howmanypetsandofwhichkinddoyouhave?'] = survey_results_df['Howmanypetsandofwhichkinddoyouhave?'].str.replace('None','0')
survey_results_df['Howmanypetsandofwhichkinddoyouhave?'] = survey_results_df['Howmanypetsandofwhichkinddoyouhave?'].str.replace(' ','')
survey_results_df['Howmanypetsandofwhichkinddoyouhave?'] = survey_results_df['Howmanypetsandofwhichkinddoyouhave?'].str.replace('):','')
survey_results_df['Howmanypetsandofwhichkinddoyouhave?'] = survey_results_df['Howmanypetsandofwhichkinddoyouhave?'].str.replace('Nopets','0')
survey_results_df['Howmanypetsandofwhichkinddoyouhave?'] = survey_results_df['Howmanypetsandofwhichkinddoyouhave?'].fillna('0')
survey_results_df['Whatisyourfavoritetypeoffood?'] = survey_results_df['Whatisyourfavoritetypeoffood?'].str.replace(' ','')
survey_results_df['Whatisyourfavoritetypeoffood?'] = survey_results_df['Whatisyourfavoritetypeoffood?'].str.replace(' ','')
survey_results_df['Whatisyourfavoritetypeoffood?'] = survey_results_df['Whatisyourfavoritetypeoffood?'].str.replace(',','')
survey_results_df['Whatisyourfavoritetypeoffood?'] = survey_results_df['Whatisyourfavoritetypeoffood?'].str.replace('/','').str.replace('.','')
```
