```python
df_clothes = pd.DataFrame(data = [['green','M',10,'shirt'],['red','M',60,'sweatshirt'],['black','S',120,'shoes']])
df_clothes.column['color','size','price','type']
df_clothes.head()
```

```python
#can convert categorical values to numerical 
#size is ordinal, so it can be easily encoded
size_map = {'S':1,'M':2,'L':3}
df_clothes['size'] = df_clothes['size'].map(size_map)
df_clothes
```
